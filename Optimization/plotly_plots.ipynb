```python
import plotly.graph_objects as go
import numpy as np

# Generate example real data points (sparser)
real_x = np.linspace(0, 10, 15)  # Fewer points for real data
real_y = np.sin(real_x)  # Main line values

# Generate the line with fewer points
line_x = np.linspace(0, 10, 30)  # Fewer points for line
line_y = np.sin(line_x)  # Main line values
upper_bound = line_y + 0.2  # Upper bound
lower_bound = line_y - 0.2  # Lower bound

# Define color palette
main_line_color = "blue"
shade_color = "rgba(0, 0, 255, 0.2)"
point_color = "darkblue"  # Matching the blue theme

# Create the plot
fig = go.Figure()

# Add the main line (less dense)
fig.add_trace(go.Scatter(x=line_x, y=line_y, mode='lines', name='Main Line', line=dict(color=main_line_color)))

# Add the shaded region for upper and lower bounds
fig.add_trace(go.Scatter(x=np.concatenate([line_x, line_x[::-1]]),
                         y=np.concatenate([upper_bound, lower_bound[::-1]]),
                         fill='toself',
                         fillcolor=shade_color,
                         line=dict(color='rgba(255,255,255,0)'),
                         name='Uncertainty Region'))

# Add real data points (matching color palette)
fig.add_trace(go.Scatter(x=real_x, y=real_y, mode='markers', name='Real Data',
                         marker=dict(color=point_color, size=8, opacity=0.8)))

# Update layout (remove grid, simplify appearance)
fig.update_layout(title='Line Plot with Upper/Lower Bounds and Real Data Points',
                  xaxis_title='X-axis',
                  yaxis_title='Y-axis',
                  showlegend=True,
                  xaxis=dict(showgrid=False),
                  yaxis=dict(showgrid=False))

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Generate example 3D data
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Example 3D function

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale="Viridis")])

# Update layout for better visualization
fig.update_layout(title='3D Surface Plot',
                  scene=dict(xaxis_title='X Axis',
                             yaxis_title='Y Axis',
                             zaxis_title='Z Axis'))

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Generate example 3D scatter data
num_points = 100
x = np.random.uniform(-5, 5, num_points)
y = np.random.uniform(-5, 5, num_points)
z = np.sin(np.sqrt(x**2 + y**2))  # Example function for Z values

# Create the 3D scatter plot
fig = go.Figure()

# Add scatter points
fig.add_trace(go.Scatter3d(x=x, y=y, z=z,
                           mode='markers',
                           marker=dict(size=5, color=z, colorscale='Viridis', opacity=0.8)))

# Update layout for better visualization
fig.update_layout(title='3D Scatter Plot',
                  scene=dict(xaxis_title='X Axis',
                             yaxis_title='Y Axis',
                             zaxis_title='Z Axis'))

# Show the plot
fig.show()
```

```python
import plotly.figure_factory as ff
import numpy as np

# Generate example data
data = np.random.normal(loc=5, scale=2, size=1000)  # Normal distribution with mean=5, std=2

# Create the histogram
fig = ff.create_distplot([data], group_labels=["Sample Data"], show_hist=True, colors=["blue"])

# Update layout for better visualization
fig.update_layout(title="Histogram of Sample Data",
                  xaxis_title="Value",
                  yaxis_title="Density",
                  showlegend=True)

# Show the plot
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate example data
data = np.random.normal(loc=5, scale=2, size=1000)  # Normal distribution with mean=5, std=2

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')

# Add labels and title
plt.title("Histogram of Sample Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```
